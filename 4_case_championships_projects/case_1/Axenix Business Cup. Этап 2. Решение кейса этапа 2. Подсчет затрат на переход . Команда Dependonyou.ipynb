# Команда Dependonyou


```python
from datetime import datetime, timedelta

GAP_ANALYSIS = {
    "АБС": 6500,
    "Процессинг": 4600,
    "ДБО": 2400,
    "Кредитный конвейер": 3700,
    "Отчетность и аналитика": 900,
}

VENDOR_RATES = {
    "АБС": 0.1,
    "Процессинг": 0.1,
}

STAFF = {
    "Нева Банк": 159,
    "Сибирский Альянс Банк": 80,
}

ROLES = {
    "Разработчик АБС": {
        "ставка": 9500,
        "производительность": 1.5,
        "ограничение": 6,
    },
    "Разработчик Процессинга": {
        "ставка": 7800,
        "производительность": 1.5,
        "ограничение": 6,
    },
    "Релиз-менеджер": {
        "ставка": 3200,
        "производительность": 1,
    },
    "Devops инженер": {
        "ставка": 4400,
        "производительность": 1,
    },
    "Специалист инфраструктуры": {
        "ставка": 6800,
        "производительность": 1,
    },
    "Архитектор": {
        "ставка": 7900,
        "производительность": 1,
    },
    "Проджект менеджер": {
        "ставка": 4900,
        "производительность": 1,
    },
}


def get_productivity(role, bank, in_team=True):
    if in_team:
        return ROLES[role]["производительность"]
    else:
        if role in ["Разработчик АБС", "Разработчик Процессинга"]:
            return ROLES[role]["производительность"] * 0.9
        else:
            return ROLES[role]["производительность"] * 0.75


def get_work_time(role, bank, duration):
    if role in ["Разработчик АБС", "Разработчик Процессинга"]:
        if duration > 2:
            return duration - 2
        else:
            return 0
    else:
        return duration


def calculate_development_costs(gap_analysis, staff, roles, vendor_rates, bank):
    total_costs = {}
    for system, gap in gap_analysis.items():
        system_costs = 0
        for role in roles:
            if role in ["Разработчик АБС", "Разработчик Процессинга"]:
                if system in ["АБС", "Процессинг"]:
                    num_employees = min(staff, int(roles[role]["ограничение"]))
                    system_costs += num_employees * roles[role]["ставка"] * get_work_time(role, bank, gap) * get_productivity(role, bank)
                else:
                    num_employees = staff * 0.6
                    system_costs += num_employees * roles[role]["ставка"] * get_work_time(role, bank, gap) * get_productivity(role, bank, in_team=False)
            else:
                num_employees = staff
                system_costs += num_employees * roles[role]["ставка"] * gap * get_productivity(role, bank)
        if system in ["АБС", "Процессинг"]:
            system_costs += system_costs * vendor_rates[system]
        total_costs[system] = system_costs
    return total_costs


def calculate_migration_costs(gap_analysis):
    migration_costs = {}
    for system, gap in gap_analysis.items():
        migration_costs[system] = gap * 0.4
    return migration_costs


def calculate_total_costs(development_costs, migration_costs):
    total_costs = {}
    for system in development_costs:
        if system in migration_costs:
            total_costs[system] = development_costs[system] + migration_costs[system]
    return total_costs


development_costs = calculate_development_costs(GAP_ANALYSIS, STAFF["Нева Банк"], ROLES, VENDOR_RATES, "Нева Банк")
migration_costs = calculate_migration_costs(GAP_ANALYSIS)
total_costs = calculate_total_costs(development_costs, migration_costs)

for system in development_costs:
    print(f"Система: {system}")
    print(f"  - Затраты на разработку: {development_costs[system]}")
    print(f"  - Затраты на миграцию: {migration_costs[system]}")
    print(f"  - Общие затраты: {total_costs[system]}")
```
