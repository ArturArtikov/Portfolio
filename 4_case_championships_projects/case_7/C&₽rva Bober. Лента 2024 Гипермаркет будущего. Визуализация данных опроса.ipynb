```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/drive/MyDrive/Opros.xlsx')
data.head()
```

```python
data['sex'] = data['17. Укажите ваш пол:']
```

```python
sex = data['sex'].value_counts().index.to_list()
count = data['sex'].value_counts().to_list()
explode = [0.1, 0, 0]
```

```python
print(sex, count, sep='\n')
```

```python
fig, axes = plt.subplots(figsize=(20, 9));
plt.pie(count, labels=sex, colors=['#FFB600', '#003594', '#00BF6F'], autopct='%1.1f%%', explode=explode);
plt.title('Соотношение полов в опросе удовлетворенности магазином "ЛЕНТА"');
```

```python
data['age'] = data['16. Укажите ваш возраст:']
age = data['age'].value_counts().index.to_list()
count_age = data['age'].value_counts().to_list()
explode_age = [0.1, 0, 0, 0]

# 16. Укажите ваш возраст:
print(age, count_age)
```

```python
fig, axes = plt.subplots(figsize=(20, 9));
plt.pie(count_age, labels=age, colors=['#FFB600', '#003594', '#00BF6F',  '#671E75'], autopct='%1.1f%%', explode=explode_age);
plt.title('Соотношение возростов в опросе удовлетворенности магазином "ЛЕНТА"');
```
